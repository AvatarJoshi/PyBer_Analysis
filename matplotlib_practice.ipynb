```python
# Add dependencies
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="^", color="green", linewidth=2, label="Boston")

# Axes Labels
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set Limits
plt.ylim(0, 45)
plt.xlim(0, 5)
# Create a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis, y_axis, label="Boston", color="green")

plt.title("PyBer Fare by Month")

plt.xlabel("Date")
plt.ylabel("Fare($)")


plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis, label="Boston")
plt.legend()
plt.xlabel("Fare($)")
```

```python
# Create the plot with ax.plot()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
plt.legend("Fare($)")
plt.xlabel("Fare($)")
```

```python
# Creates list of larger values to be used for proportionally increasing size of scatter points
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, label="Chicago", color="red", marker="o", s=y_axis_larger)
plt.ylabel("Fare($)")
plt.legend()

```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis_larger, color="darkcyan", edgecolors="black", alpha=0.7, label="Boston")
ax.legend()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

plt.pie(y_axis, labels=x_axis, explode=explode_values, autopct="%.1f%%", colors=colors)
        
plt.show()
```

```python
explode_values = (0, 0, 0.3, 0, 0, 0, 0.3, 0, 0, 0, 0, 0)

fig, ax = plt.subplots()
ax.pie(y_axis, labels = x_axis, radius= 2, explode = explode_values, shadow=True, autopct="%.1f%%")
plt.show()
```
